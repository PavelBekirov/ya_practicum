## Алгоритм преобразования

**Алгоритм**

1. Сгенерируем случайным образом обратимую матрицу $Z$. Ее определитель ≠ 0 ($|Z| = det(Z)≠ 0$)

2. Умножаем  матрицу $X$ на  $Z$. Получим новую матрицу $Y = XZ$

3.  Докажем, что качетсво линейной регрессии не изменится.  

**Обоснование**

1. Наша случайно сгенерированная матрица $Z$ должна иметь размерность 4x4, так как наша матрица признаков $X$ имеет 4 столбцов ( и 5000 строк, но это не важно. главное, что умножаем матрицу $X$ с 4-мя столбцами на матрицу $Z$ (4х4)).1 столбец у таргета, итого 5 столбцов

2. Находим определитель квадратной матрицы $Z$. Он не должен равняться 0. $|Z| = det(Z)≠ 0$. То есть наша матрица $Z$ обратимая. 
3. Получаем матрицу $Y = XZ$

## Проверка алгоритма

На простой линейной регрессии стандартизация не окажет никакого влияния, поскольку все изменения переменной можно нивелировать за счет изменения коэффициента:

$y = \alpha + \beta X$ - без стандартизации

$y = \alpha_{st} + \beta_{st} \frac{X-mean}{std} = (\alpha_{st} - \frac{\beta_{st}mean}{std}) + (\frac{\beta_{st}}{std}) X$ - со стандартизацией

При этом $\alpha = (\alpha_{st} - \frac{\beta_{st}mean}{std})$ и $\beta = (\frac{\beta_{st}}{std})$.


### До преобразований

```python
features_train, features_test, target_train, target_test = train_test_split(features,
                                                                            target, 
                                                                            test_size=0.25, 
                                                                            random_state= RANDOM_STATE)
```

```python
# Без масштабирования
model = LinearRegression()
model.fit(features_train, target_train) # обучаем на train
predictions = model.predict(features_test) # предсказания на test`e
quality = r2_score(target_test, predictions)
print('метрика R2 до масштабирования', quality)
```

```python
# Промасштабируем признаки, чтобы убедиться, что метрика не изменится
scaler = StandardScaler()

scaler.fit(features_train)
scaler_features_train = scaler.transform(features_train)
scaler_features_test = scaler.transform(features_test)

# таргеты нельзя стандартизировать
```

```python
# С масштабированием
model_scaler = LinearRegression()
model_scaler.fit(scaler_features_train, target_train)
quality_scaler = r2_score(target_test, model_scaler.predict(scaler_features_test))
print('метрика R2 после масштабирования', quality_scaler)
```

### После преобразований

```python
# Создадим функцию нашего алгоритма преобразования матрицы признаков
def transform(features):
    np.random.seed(98765) # случайным образом генерируем мутрицу
    matrix = np.random.randint(10, 20, (4,4)) # задаем размерность 4х4 и генерируем числа в матрице от 10 до 20
    det = np.linalg.det(matrix) # применим функцию для вычисления определителя (детерминанта) матрицы
    
    while det == 0: # в цикле while перед очередным шагом происходит проверка условия
        np.random.seed(56789)
        matrix = np.random.randint(10, 20, (4,4))
        det = np.linalg.det(matrix)
    features_transform = features @ matrix # умножаем матрицу признаков на случайно сгенерированную квадратную матрицу
    return features_transform, matrix
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score


data_train = pd.read_csv('train_data_n.csv')
features_train = data_train.drop(['target'], axis=1)
target_train = data_train['target']

data_test = pd.read_csv('test_data_n.csv')
features_test = data_test.drop(['target'], axis=1)
target_test = data_test['target']


class SGDLinearRegression:
    def __init__(self, step_size, epochs, batch_size):
        self.step_size = step_size
        self.epochs = epochs
        self.batch_size = batch_size

    def fit(self, train_features, train_target):
        X = np.concatenate((np.ones((train_features.shape[0], 1)), train_features), axis=1)
        y = train_target
        w = np.zeros(X.shape[1])

        for _ in range(self.epochs):
            batches_count = target_train.shape[0]
            for i in range(batches_count):
                begin = i * self.batch_size
                end = (i + 1) * self.batch_size
                X_batch = X[begin:end, :]
                y_batch = y[begin:end]
                gradient = (2 * X_batch.T.dot(X_batch.dot(w) - y_batch)) / X_batch.shape[0]
                w -= self.step_size * gradient

        self.w = w[1:]
        self.w0 = w[0]
        self.batches_count = batches_count
    def predict(self, test_features):
        return test_features.dot(self.w) + self.w0

model = SGDLinearRegression(0.01, 10, 100)
model.fit(features_train, target_train)
pred_train = model.predict(features_train)
pred_test = model.predict(features_test)
print(r2_score(target_train, pred_train).round(5))
print(r2_score(target_test, pred_test).round(5))
```
