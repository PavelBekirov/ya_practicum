#### Столбец "ceiling_height"

- Тип: количественный
- Причина: человеческий фактор, не все указывают высоту потолков.

Гистограмма строится с масштабом до значения 100, а значит в данных есть ошибки и аномалии.
Попробуем отсортировать столбец по убыванию:

```python
data['ceiling_height'].sort_values(ascending=False)
```

Тут явная ошибка порядком - не все потолки в метрах. Найдем, откуда значения начинаются не в метрах:

```python
data.query('ceiling_height > 4').sort_values(by='ceiling_height', ascending=False).head(5)
```

В диапазоне от 20 до 100 метров явная ошибка - эти значения нужно разделить на 10 чтобы получить верные.
Также есть аномалии с высотами в 10 и 14 метров. Пока оставим их без изменений.

```python
# Меняем значения от 20 до 100
data.loc[(data['ceiling_height'] >= 20), 'ceiling_height'] = data['ceiling_height'] / 10
```

```python
# Супер! Обновим гистограмму:
data['ceiling_height'].hist(bins=70, figsize=(8,3))
```

Теперь сочтем все что выше 4 метров и ниже 2-х - это выбросы. Сначала удалим их. Потом заменим все NaN на медиану,чтобы выбросы и потолки в 8-14 метров (единичные случаи) не повлияли на выборку.

```python
len(data[(data['ceiling_height'] > 4) | (data['ceiling_height'] < 2.5)])
```

```python
data.drop(data.loc[(data['ceiling_height'] > 4) | (data['ceiling_height'] < 2.5)].index, inplace=True)
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
data['ceiling_height'].hist(bins=70, figsize=(8,3))
```

```python
data.drop(data.loc[data['floors_total'] > 30].index, inplace=True)
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
data['floors_total'].hist(bins=30, figsize=(8,3))
```

##### Проверим на неявные дубликаты:
- Приведем все к нижнему регистру
- Изменим все ё на е
- Приведем к универсальным записям, оставив деревни, поселки, cела и садовые товарищества.

```python
# кол-во уникальных значений до обработки
print(len(data.locality_name.sort_values().unique()))

data['locality_name'] = data['locality_name'].str.lower()
data['locality_name'] = data['locality_name'].str.replace('ё','е')
data['locality_name'] = data['locality_name'].str.replace(' городского типа','')
data['locality_name'] = data['locality_name'].str.replace('городской ','')
data['locality_name'] = data['locality_name'].str.replace('коттеджный ','')
data['locality_name'] = data['locality_name'].str.replace('садоводческое некоммерческое товарищество','садовое товарищество')
#data['locality_name'] = data['locality_name'].str.replace('поселок','пос.')
print(data.locality_name.sort_values().unique())

# кол-во уникальных значений после обработки
print(len(data.locality_name.sort_values().unique()))
```

Мы убрали 43 дубликата!

#### Проверка автоматически собранных данных

Как было сказано в ТЗ - у нас 2 типа данных. Сгруппируем данные по локации и посмотрим пропуски в нас пунктах.

```python
pd.options.display.max_rows = 50
pt = data.pivot_table(
    index='locality_name', 
    values=['airports_nearest', 'cityCenters_nearest', 'parks_nearest', 'ponds_nearest'],
    aggfunc=[lambda x: x.isna().sum(), 'count'])
pt.columns=(
    ['airports_nearest_nans', 'cityCenters_nearest_nans', 'parks_nearest_nans', 'ponds_nearest_nans',
     'airports_nearest_counts', 'cityCenters_nearest_counts', 'parks_nearest_counts', 'ponds_nearest_counts']
)
pt.sort_values(by='airports_nearest_counts', ascending=False)
```

Видим, что пропуски одинаковые. Явно выбивается СПБ - в нем пропуски в парках и водоемах составляют половину выборки.
Мы поняли, что пропуски связанв с СПБ, а значит расстояния считались практически только для города. Учтем это, но пока оставим пропуски как NaN.

#### Приведение float64 в int

Проверим все столбцы на наличие дробных чвастей - если их нет, то можно перевести эти столбцы в int

```python
print("last_price", data.query('last_price % 1 != 0').last_price.count())
print('total_area', data.query('total_area % 1 != 0').last_price.count())
print('ceiling_height', data.query('ceiling_height % 1 != 0').last_price.count())
print('floors_total', data.query('floors_total % 1 != 0').last_price.count())
print('balcony', data.query('balcony % 1 != 0').last_price.count())
print('parks_around3000', data.query('parks_around3000 % 1 != 0').last_price.count())
print('ponds_around3000', data.query('ponds_around3000 % 1 != 0').last_price.count())
print('days_exposition', data.query('days_exposition % 1 != 0').days_exposition.sum())
```

last_price, floors_total, balcony, parks_around3000, ponds_around3000, days_exposition можно перевести integer.

```python
data['last_price'] = data['last_price'].astype('int')
data['floors_total'] = data['floors_total'].astype('int')
data['balcony'] = data['balcony'].astype('int')
data['parks_around3000'] = data['parks_around3000'].astype('int')
data['ponds_around3000'] = data['ponds_around3000'].astype('int')
data['days_exposition'] = data['days_exposition'].astype('int')
display(data.dtypes)
```

## Посчитайте и добавьте в таблицу новые столбцы

### Цена одного квадратного метра

```python
# Так как я привел общий ценник к int, то и за 1 кв м тоже приведу к int
data['m2_price'] = (data['last_price'] / data['total_area']).astype(int)
data.head(3)
```

### День недели, месяц и год публикации объявления

```python
data['weekday'] = pd.DatetimeIndex(data['first_day_exposition']).weekday
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year
data.head(3)
```

### Тип этажа квартиры

```python
import numpy as np
data['floot_type'] = np.where(data.floor == data.floors_total, 'последний', '')
data['floot_type'] = np.where(data.floor == 1, 'первый', data['floot_type'])
data['floot_type'] = np.where(~data.floot_type.isin(['первый','последний']), 'другой', data['floot_type'])
data.head(5)
```

### Расстояние до центра в км

```python
data['cc_nearest_km'] = (
    np.where(data.cityCenters_nearest != 'NaN', (data['cityCenters_nearest'] / 1000).round(), data.cityCenters_nearest))
data
```

#### Зависимость от общей площади

```python
# Напишем функцию для построения от 2-х зависимостей с красивыми графиками и возможностью перевернуть координаты.
def plotting(a,b='last_price', c='not_reverse', d='median'):
    if c != 'reverse':
        data_graph = data.pivot_table(index=a, values=b, aggfunc = d).sort_values(by=b)
        data_graph[b] = data_graph[b] / 1_000_000
        x = list(data_graph[b])
        y = data_graph.index # потому кта эти и есть лист зависимого значения
        fig, axs = plt.subplots(1, 2, figsize=(11, 3), sharey=True)
        axs[0].scatter(x, y)
        axs[0].set_ylabel(f'{b}, m2')
        axs[0].set_xlabel('Price, mln rub')
        axs[1].set_ylabel(f'{b}, m2')
        axs[1].set_xlabel('Price, mln rub')
        axs[1].plot(x, y)
        fig.suptitle(f'Графики зависимости {a} и {b}')
    else:
        data_graph = data.pivot_table(index=a, values=b, aggfunc = d).sort_values(by=a)
        data_graph[b] = data_graph[b] / 1_000_000
        y = list(data_graph[b])
        x = data_graph.index # потому кта эти и есть лист зависимого значения
        fig, axs = plt.subplots(1, 2, figsize=(11, 3), sharey=True)
        axs[0].scatter(x, y)
        axs[0].set_xlabel(f'{b}, m2')
        axs[0].set_ylabel('Price, mln rub')
        axs[1].set_xlabel(f'{b}, m2')
        axs[1].set_ylabel('Price, mln rub')
        axs[1].plot(x, y)
        fig.suptitle(f'Графики зависимости {b} от {a}')
    
plotting('total_area')
```

Цена растет вместе с площадью. Но не все так однозначно, так как и другие факторы влияют.
