### Работа с пропусками

```python
data.isna().sum()
```

#### Столбец "ceiling_height"

- Тип: количественный
- Причина: человеческий фактор, не все указывают высоту потолков.

Гистограмма строится с масштабом до значения 100, а значит в данных есть ошибки и аномалии.
Попробуем отсортировать столбец по убыванию:

```python
data['ceiling_height'].sort_values(ascending=False)
```

Тут явная ошибка порядком - не все потолки в метрах. Найдем, откуда значения начинаются не в метрах:

```python
data.query('ceiling_height > 4').sort_values(by='ceiling_height', ascending=False).head(5)
```

В диапазоне от 20 до 100 метров явная ошибка - эти значения нужно разделить на 10 чтобы получить верные.
Также есть аномалии с высотами в 10 и 14 метров. Пока оставим их без изменений.

```python
# Меняем значения от 20 до 100
data.loc[(data['ceiling_height'] >= 20), 'ceiling_height'] = data['ceiling_height'] / 10
```

```python
# Супер! Обновим гистограмму:
data['ceiling_height'].hist(bins=70, figsize=(8,3))
```

Теперь сочтем все что выше 4 метров - это выбросы. Сначала удалим их. Потом заменим все NaN на медиану,чтобы выбросы и потолки в 8-14 метров (единичные случаи) не повлияли на выборку.

```python
data.drop(data.loc[data['ceiling_height'] > 4].index, inplace=True)
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
data['ceiling_height'].hist(bins=70, figsize=(8,3))
```

#### Столбец "floors_total"

- Тип: количественный
- Причина: не все помнят скольки этажный у него дом, не заполняют это поле. Или дом не имеет этажность

Хорошо, что всего 86 пропусков, пойдем пока теми же шагами:
- Смотрим распределение и аномалии
- Исправляем аномалии
- Заполняем пропуски медианой

```python
# Пропусков не много, построим гистограмму и проверим выбросы:
data['floors_total'].hist(bins=30, figsize=(8,3))
```

Видим, что есть пара очень высоких домов, но в остальном все стандартно - 5-ны и 9-ны выбиваются. Думаю можно заменять пропуски медианой после удаления выбросов (30 этажей и выше).

```python
data.drop(data.loc[data['floors_total'] > 30].index, inplace=True)
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
data['floors_total'].hist(bins=30, figsize=(8,3))
```

#### Столбец "living_area"

- Тип: количественный
- Причина: По какой то причине жидая площадь не указана.

Не сказать что пропусков очень много, поэтому применим шаги:

- Исправляем аномалии

```python
data.living_area.hist(bins=100, figsize=(8,3))
```

Похоже на распределение Пуассона (выбросы более 100 не смотрю на гистограмме чтобы понять причину 3 пиков). 
В итоге причина пиков - однушка, двушка, трешка типовых застроек, как мне кажется все правдоподобно.
Больше 100 удалим.
Думаю идеально для медианы в пропусках чтобы не искажали пентхаусы с 400 кв м))

```python
data.drop(data.loc[data['living_area'] > 100].index, inplace=True)
data['living_area'] = data['living_area'].fillna(data['living_area'].median())
data.living_area.hist(bins=100, figsize=(8,3))
```
