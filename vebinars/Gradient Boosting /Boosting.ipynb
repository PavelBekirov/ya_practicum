```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.tree import DecisionTreeRegressor
```

```python
X = np.arange(0, 6 * np.pi, 0.01)

y = np.sin(X) + 1.5 * np.random.rand(len(X))
y[::20] = 5 * np.random.rand(len(y[::20])) - 2.5
#y = np.sin(X)

plt.scatter(X, y, s=1)

X = X.reshape(-1, 1)
```

```python
clf = DecisionTreeRegressor(max_depth=3)

clf.fit(X, y)
```

```python
plt.scatter(X.ravel(), y, s=1)
y_hat = clf.predict(X)
plt.plot(X.ravel(), y_hat, c="orange", lw=3)
```

```python
y1 = y - y_hat

plt.scatter(X.ravel(), y1, s=1)
```

```python
clf1 = DecisionTreeRegressor(max_depth=3)
clf1.fit(X, y1)
```

```python
plt.scatter(X.ravel(), y1, s=1)
y_hat = clf1.predict(X)
plt.plot(X.ravel(), y_hat, c="orange", lw=3)
```

```python
y2 = y1 - y_hat

plt.scatter(X.ravel(), y2, s=1)
```

```python
clf2 = DecisionTreeRegressor(max_depth=3)
clf2.fit(X, y2)
```

```python
plt.scatter(X.ravel(), y2, s=1)
y_hat = clf2.predict(X)
plt.plot(X.ravel(), y_hat, c="orange", lw=3)
```

```python
y3 = y2 - y_hat

plt.scatter(X.ravel(), y3, s=1)
```

```python
plt.scatter(X.ravel(), y, s=1)
y_hat_final = clf.predict(X) + clf1.predict(X) + clf2.predict(X) 
plt.plot(X.ravel(), y_hat_final, c="orange", lw=3)
```
